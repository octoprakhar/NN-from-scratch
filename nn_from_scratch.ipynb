```python
## There can be four input or neurons in the input layer [0,0], [0,1], [1,0], [1,1] as 1,2,3,4, with all some percentage for activation like [0.22,0.34,]
## Each hidden layer will have 6 neaurons


class XOR_NN_Sratch:
    def __init__(self,inputLayerActivations,learning_rate=0.1):
        self.inputLayerActivations = inputLayerActivations
        self.learning_rate = learning_rate
        self.W1 = np.random.randn(2,6)
        self.b1 = np.random.randn(1,6)

        self.W2 = np.random.randn(6,6)
        self.b2 = np.random.randn(1,6)

        self.W3 = np.random.randn(6,1)
        self.b3 = np.random.randn(1,1)

    def sigmoid(self, z):
        return 1/(1+np.exp(-z))
    
    def forward(self, x):
        self.a0 = np.asarray(x).reshape(1,2)
        self.z1 = self.a0 @ self.W1 + self.b1
        self.a1 = self.sigmoid(self.z1)

        self.z2 = self.a1 @ self.W2 + self.b2
        self.a2 = self.sigmoid(self.z2)

        self.z3 = self.a2 @ self.W3 + self.b3
        self.a3 = self.sigmoid(self.z3)
        return self.a3
    
    def backward(self,x,y):
        y = np.asarray(y).reshape(1,1)

        # forward pass already done before calling this
        delta_3 = (self.a3-y) * self.a3*(1-self.a3) 

        dw3 = self.a2.T @ delta_3
        db3 = delta_3

        self.W3 -= (dw3 * self.learning_rate)
        self.b3 -= (db3 * self.learning_rate)

        delta_2 = (delta_3 @ self.W3.T) * self.a2*(1-self.a2)
        dw2 = self.a1.T @ delta_2
        db2 = delta_2

        self.W2 -= (dw2 * self.learning_rate)
        self.b2 -= (db2 * self.learning_rate)

        delta_1 = (delta_2 @ self.W2.T) * self.a1*(1-self.a1)
        dw1 = self.a0.T @ delta_1
        db1 = delta_1

        self.W1 -= (dw1 * self.learning_rate)
        self.b1 -= (db1 * self.learning_rate)

    def predict(self, x):
        return self.forward(x)
```

```python
# Training data for XOR
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

# Initialize network
xor_nn = XOR_NN_Sratch(inputLayerActivations=None, learning_rate=0.1)

# Training loop
epochs = 10000
for epoch in range(epochs):
    loss = 0
    for i in range(len(X)):
        y_hat = xor_nn.forward(X[i])              # Forward pass
        xor_nn.backward(X[i], y[i])               # Backward pass (update weights)
        loss += np.mean((y[i] - y_hat)**2)        # Calculate MSE for this sample

    if epoch % 1000 == 0:  
        print(f"Epoch {epoch}, Loss: {loss/4:.4f}")   # average loss over 4 samples
```
